Binning
Binning/converting numerical age to categorical Variable

Map:

child : 0
young : 1
adult : 2
mid-age : 3
senior : 4


```python
for ds in train_test:
    ds.loc[ds['Age'] <=16 ,'Age'] =0,
    ds.loc[(ds['Age']>16) & (ds['Age'] <= 26), 'Age'] = 1,
    ds.loc[(ds['Age']>26) & (ds['Age'] <= 36), 'Age'] = 2,
    ds.loc[(ds['Age']>36) & (ds['Age'] <= 62), 'Age'] = 3,
    ds.loc[ds['Age']>62 , 'Age'] =4
```

```python
#We can count total family mamber 
for dataset in train_test:
    dataset['F_size'] = dataset['SibSp']+dataset['Parch']+1
```

```python
# fill fare missing with median fare for each class

train_data['Fare'].fillna(train_data.groupby('Pclass')['Fare'].transform("median"),inplace=True)
test_data['Fare'].fillna(test_data.groupby('Pclass')['Fare'].transform("median"),inplace=True)
```

```python
# mapping in Fare

for dataset in train_test:
    dataset.loc[dataset['Fare'] <= 17, 'Fare'] = 0,
    dataset.loc[(dataset['Fare'] > 17) & (dataset['Fare'] <= 30), 'Fare'] = 1,
    dataset.loc[(dataset['Fare'] > 30) & (dataset['Fare'] <= 100), 'Fare'] = 2,
    dataset.loc[dataset['Fare'] >= 100, 'Fare'] = 3
```
